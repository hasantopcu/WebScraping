```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
import math
```

```python
from_host_istanbul = pd.read_csv('istanbul_sahibinden.csv')
print('Tuple counts for İstanbul: ' , len(from_host_istanbul))
from_host_ankara = pd.read_csv('ankara_sahibinden.csv')
print('Tuple counts for Ankara:   ' , len(from_host_ankara))
from_host_izmir =  pd.read_csv('izmir_sahibinden.csv')
print('Tuple counts for İzmir:    ' , len(from_host_izmir))
from_host_adana = pd.read_csv('adana_sahibinden.csv')
print('Tuple counts for Adana:    ' , len(from_host_adana))
```

```python
x = from_host_istanbul[['Fiyat','Ilce']]
for i in range(0 , len(x['Fiyat'])):
    x['Fiyat'][i] =  int(x['Fiyat'][i].replace('TL' , '').replace('.' , ''))
```

```python
plt.scatter(x['Ilce'] , x['Fiyat']  ,  c = x['Fiyat'] ,  marker='v' , cmap='Greens' , edgecolor='black' , linewidth=1)
cbar = plt.colorbar()
cbar.set_label('İstanbul monthly average rent (TL)')
plt.xticks(rotation=90)
plt.title('Monthly average rent is 2415 TL')
plt.show()
```

```python
y = from_host_ankara[['Fiyat','Ilce']]
for i in range(0 , len(y['Fiyat'])):
    y['Fiyat'][i] =  int(y['Fiyat'][i].replace('TL' , '').replace('.' , ''))
    if y['Fiyat'][i] > 20000:
        y['Fiyat'][i] = y['Fiyat'][i]/12
```

```python
plt.scatter(y['Ilce'] , y['Fiyat'] , marker='v' ,  c = y['Fiyat'] , cmap='Blues' , edgecolor='black' , linewidth=1)
cbar = plt.colorbar()
cbar.set_label('Ankara monthly average rent (TL)')
plt.xticks(rotation=90)
plt.title('Monthly average rent is 1978 TL')
plt.show()
```

```python
z = from_host_izmir[['Fiyat','Ilce']]
for i in range(0 , len(z['Fiyat'])):
    z['Fiyat'][i] =  int(z['Fiyat'][i].replace('TL' , '').replace('.' , ''))
    if z['Fiyat'][i] > 10000:
        z['Fiyat'][i] =z['Fiyat'][i]/12
```

```python
plt.scatter(z['Ilce'] , z['Fiyat'] , c = z['Fiyat'] ,  marker='v' , edgecolor='black' , cmap='Reds' , linewidth=1)
cbar = plt.colorbar()
cbar.set_label('İzmir monthly average rent (TL)')
plt.xticks(rotation=90)
plt.title('Monthly average rent is 2057 TL')
plt.show()
```

```python
t = from_host_adana[['Fiyat','Ilce']]
for i in range(0 , len(t['Fiyat'])):
    t['Fiyat'][i] =  int(t['Fiyat'][i].replace('TL' , '').replace('.' , ''))
    if t['Fiyat'][i] > 4000:
        t['Fiyat'][i] = t['Fiyat'][i]/12
```

```python
plt.scatter(t['Ilce'] , t['Fiyat'], c = t['Fiyat'] ,  marker='v' , edgecolor='black' , cmap='Purples' , linewidth=1)
cbar = plt.colorbar()
cbar.set_label('Adana monthly average rent (TL)')
plt.xticks(rotation=90)
plt.title('Monthly average rent is 1099 TL')
plt.show()
```
